### 1. Load Data and Initial Analysis

```python
#Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Historical Weather Data : Direct Solar Radiation**

```python
df1 = pd.read_csv('C:/Users/Michelle Natali/Downloads/Group 4 - Data/predict-energy-behavior-of-prosumers/historical_weather.csv')
cleaned_df1 = df1.dropna()
cols_to_check_zeros = ['direct_solar_radiation']  
cleaned_df1 = cleaned_df1[(cleaned_df1[cols_to_check_zeros] != 0).all(axis=1)]

df1 = cleaned_df1
df1
```

```python
df1.info()
```

```python
#Only show desired information to save memory

desired_columns = ['datetime', 'direct_solar_radiation']
reordered_df = df1[desired_columns]
df1 = reordered_df
df1
```

```python
df1['direct_solar_radiation'] = df1['direct_solar_radiation'].astype('float32')
df1.info()
```

**Production, Consumption, Installed_capacity and Datetime**

```python
# Data from previous session (target_data_0)
df2 = pd.read_csv('C:/Users/Michelle Natali/Downloads/Group 4 - Data/predict-energy-behavior-of-prosumers/merged_production_consumption_installed_capacity_and_datetime.csv')
df2

desired_columns = ['county_name','datetime', 'consumption','installed_capacity', 'production']
reordered_df = df2[desired_columns]
df2 = reordered_df
df2
```

```python
df2['consumption'] = df2['consumption'].astype('float32')
df2['installed_capacity'] = df2['installed_capacity'].astype('float32')
df2['production'] = df2['production'].astype('float32')
df2.info()
```

```python
# Merging county name, consumption, production, installed capacity and direct solar radiation
merged_df = pd.merge(df1, df2, on=['datetime'], how='left')
merged_df
```

```python
mean_direct_solar_radiation = merged_df.groupby('county_name')['direct_solar_radiation'].mean().reset_index()
mean_direct_solar_radiation
mean_consumption = merged_df.groupby('county_name')['consumption'].mean().reset_index()
mean_consumption
mean_production = merged_df.groupby('county_name')['production'].mean().reset_index()
mean_production
mean_installed_capacity = merged_df.groupby('county_name')['installed_capacity'].mean().reset_index()
mean_installed_capacity

mean = merged_df.groupby('county_name')[['consumption','production','installed_capacity','direct_solar_radiation']].mean().reset_index()
mean
```

```python
#Show Info
mean.info()
```

```python
mean = mean.rename(columns={'county_name': 'Name'})
mean
```

```python
#Saving the file into Excel files
mean.to_csv('C:/Users/Michelle Natali/Downloads/Group 4 - Data/predict-energy-behavior-of-prosumers/merged_production_consumption_installed_capacity_datetime_solar_radiation.csv')
```
